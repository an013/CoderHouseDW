```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
dados =  pd.read_csv('base/HRDataset_v14.csv')
dados.head(2)
```

```python
mpl.style.use('bmh')
```

```python
tabela =  pd.DataFrame(dados)
#tabela  = tabela.loc[(tabela['Salary'] > 0,('Salary'))]
tabela
tabelas_ajustada = tabela.groupby(['Position']).size().reset_index(name='Total_Salarios')
tabelas_ajustada
```

```python
#Fazer um histograma com salÃ¡rios. Quais faixas salariais sÃ£o as mais populares?
sns.set()
fig, ax = plt.subplots(figsize=(20,4))
plt.xticks(rotation=90)#Legenda na vertical
ax.bar(tabelas_ajustada['Position'], tabelas_ajustada['Total_Salarios'])
ax.set_title('Salario mais Popular')
ax.set_ylabel('Quantidade de Salarios')
ax.set_xlabel('Profissao')
```

```python
#TraÃ§ar a sÃ©rie temporal correspondente Ã  data de contrataÃ§Ã£o (DateofHire)
```

```python
dados['DateofHire']= pd.to_datetime(dados['DateofHire'])

dados_date = dados.groupby(['DateofHire']).size().reset_index(name='qtd')
dados_date
```

```python
fig, ax = plt.subplots(figsize=(20,4))
ax.plot(dados_date['DateofHire'], dados_date['qtd'])

ax.grid(True, linestyle='-.')
ax.tick_params(labelcolor='r', labelsize='medium', width=3)

plt.show()
```
