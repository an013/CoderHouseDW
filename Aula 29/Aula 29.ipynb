# 2. Corrigir os target a apenas dois valores distintos ‘<=50k’ e ‘>50k’

```python
dados_geral =  pd.merge(dados_w,dados_p, on ='ID')

pd.Series(dados_geral['Target'], name='A').unique() #verificação dos valores distindos da coluna 'Target'
```

```python
dados_geral['Target'] = dados_geral['Target'].map({' >50K.':' >50K',
                                             ' <=50K.':' <=50K'},
                                              na_action=None)

 
dados_geral
```

def altera_target(valor):
    if valor =='<=50K.':
        return '<=50K'
    if valor =='>50K.':
        return '>50K'
    return valor

dados_geral['Target'] = dados_geral['Target'].apply(altera_target)
dados_geral

# 3. Calcular a proporção entre pessoas com salários >50k em relação ao total com pessoas que difiram em sexo, raça e nível educacional. 
Por exemplo, ache qual é a porcentagem de mulheres com um salário maior que 50K e compare com a dos homens. O mesmo para pessoas de diferentes raças e níveis educacionais.

```python
# Cálculo da porcentagem de homems e mulheres com salários >50k em relação ao total de pessoas.
df_Sex = dados_geral['Sex'].loc[dados_geral['Target'] == ' >50K'].value_counts()
df_Sex / 48841 * 100
```

```python
# Cálculo da porcentagem de pessoas por raça com salários >50k em relação ao total de pessoas.
df_Race = dados_geral['Race'].loc[dados_geral['Target'] == ' >50K'].value_counts()
df_Race / 48841 * 100
```

```python
# Cálculo da porcentagem de pessoas por níveis educacionais com salários >50k em relação ao total de pessoas.
df_Education = dados_geral['Education'].loc[dados_geral['Target'] == ' >50K'].value_counts()
df_Education / 48841 * 100
```
