# 2. Corrigir os target a apenas dois valores distintos â€˜<=50kâ€™ e â€˜>50kâ€™

```python
dados_geral =  pd.merge(dados_w,dados_p, on ='ID')

pd.Series(dados_geral['Target'], name='A').unique() #verificaÃ§Ã£o dos valores distindos da coluna 'Target'
```

```python
dados_geral['Target'] = dados_geral['Target'].map({' >50K.':' >50K',
                                             ' <=50K.':' <=50K'},
                                              na_action=None)

 
dados_geral
```

def altera_target(valor):
    if valor =='<=50K.':
        return '<=50K'
    if valor =='>50K.':
        return '>50K'
    return valor

dados_geral['Target'] = dados_geral['Target'].apply(altera_target)
dados_geral
